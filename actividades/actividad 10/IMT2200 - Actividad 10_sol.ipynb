## 1. Base de datos

### 1.1 Lectura y preparacin
Lea el archivo `oceanData.csv`, dejando la informacin en un DataFrame llamado `oceanData`. Luego agregue una columna llamada `decada` que agregue la dcada de esa muestra. (Recomendacin: usar la columna `Year` para sacar la dcada correspondiente).

### 1.2 Visualizacin
Para cada dcada, grafique los datos de temperatura (independiente) vs salinidad (dependiente) en un `scatter plot` que muestre la salinidad como funcin de la temperatura. No olvide que todo grfico debe tener nombres en los ejes.

```python
# lectura de datos
oceanData = pd.read_csv('oceanData.csv')
oceanData.head()
```

```python
# agregar dcada
oceanData['decada'] = np.floor(oceanData['Year']/10)*10
```

```python
# graficar datos por d'ecada
d = oceanData['decada'].unique()

maxT = max(oceanData['T_degC'])
minT = min(oceanData['T_degC'])
maxS = max(oceanData['Salnty'])
minS = min(oceanData['Salnty'])

fig = plt.figure(figsize=(15,10))
for i in d:
    oceanDec = oceanData[oceanData['decada'] == i]
    ax = fig.add_subplot(3,3,np.where(d == i)[0][0]+1)
    plt.plot(oceanDec['T_degC'], oceanDec['Salnty'], '.', markersize=2)
    ax.set_xlim([minT, maxT])
    ax.set_ylim([minS, maxS])
    ax.set_xlabel('Temperatura')
    ax.set_ylabel('Salinidad [%i]'%(i))
```

## 2. Relaciones

Queremos ver si la relacion entre las variables medidas ha ido cambiando en el tiempo. En la Interrogacin vimos que el mejor modelo era el multilineal, as que aprovecharemos eso para entender cmo cambia la relacin entre las variables en el tiempo. ### 2.1 Modelos anuales
Para cada ao, haga un modelo de regresin multilineal que considere la `Salinidad` como variable dependiente y las mediciones de `Temperatura`, `Profundidad` y `Densidad` como las variables independientes. No necesita separar los datos entre prueba y validacin para este ejercicio.

En cada modelo, calcule el r, el RMSE y almacene los valores de $\beta_0$, $\beta_1$, $\beta_2$ y $\beta_3$ (los coeficientes de la regresin).

### 2.2 Visualizacin
Haga los siguientes grficos:
- Un nico grfico que muestre, para cada ao, el valor de los coeficientes de la regresin multilineal ($\beta_0$, $\beta_1$, $\beta_2$ y $\beta_3$). Se mantienen estables a lo largo de los aos?
- Un grfico que muestre el valor de r para cada ao.
- Un grfico que muestre el valor del RMSE para cada ao. Cambia el nivel de error a lo largo de los aos?

```python
# librerias necesarias
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# aos en los datos
years = oceanData['Year'].unique()

# listas para almacenar las mediciones solicitadas
r2 = []
rmse = []
beta0 = []
beta1 = []
beta2 = []
beta3 = []

# armar un modelo por ao
for y in years:
    oceanYear = oceanData[oceanData['Year'] == y]
    
    y_train = np.array(oceanYear['Salnty'])
    X_train = np.array(oceanYear[['T_degC', 'Depthm', 'STheta']])

    # creamos y entrenamos el modelo
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)

    # calculo de funcionamiento
    y_pred_train = linreg.predict(X_train)
    r2_lineal_train = linreg.score(X_train, y_train)
    rmse_train = mean_squared_error(y_train, y_pred_train)
    
    r2.append(r2_lineal_train)
    rmse.append(rmse_train)
    
    beta0.append(linreg.intercept_)
    beta1.append(linreg.coef_[0])
    beta2.append(linreg.coef_[1])
    beta3.append(linreg.coef_[2])
```

```python
# graficar los valores de beta para cada ao
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(111)
plt.plot(years, beta0, '.-', markersize=2, label=r'$\beta_0$')
plt.plot(years, beta1, '.-', markersize=2, label=r'$\beta_1$')
plt.plot(years, beta2, '.-', markersize=2, label=r'$\beta_2$')
plt.plot(years, beta3, '.-', markersize=2, label=r'$\beta_3$')
ax.set_xlabel('Year')
ax.set_ylabel('Value')
ax.legend()
```

```python
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(111)
plt.plot(years, r2, 'o-', markersize=4)
ax.set_xlabel('Year')
ax.set_ylabel('r^2')
```

```python
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(111)
plt.plot(years, rmse, 'o-', markersize=4)
ax.set_xlabel('Year')
ax.set_ylabel('RMSE')
```
