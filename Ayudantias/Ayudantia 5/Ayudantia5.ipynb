## Extracción

### Tipos de datos

**Estructurados:** Los datos estructurados son información que ha sido formateada y transformada en un modelo de datos bien definido, tenemos ejemplos como  Excel, bases de datos relacionales, que se estructuran con columnas y filas


**Semi-Estructurados:** Existen etiquetas que nos permiten tener jerarquía parcial a los datos, JSON, XML


**No estructurados:** No hay jerarquia interna, podemos pensar que se encuentra la información de una pagína web, comentarios de redes sociales, respuestas de encuestas, etc.

**La base de datos de aplicaciones que trabajamos en esta ayudantia corresponde a una base de datos estructurada.**

**Data Warehouse:** Estructura que conecta múltiples bases de datos, con el objetivo de proever de información actualizada, relevante y estratégica para una empresa u organización.


**Data Lake:** Estructura que almacená diferentes tipos de datos, de diferentes fuentes.





Entre ambos términos, su gran diferencia es que en el Data Warehouse se almacenan datos procesados y refinados, en cambio en un Data Lake se almacena data cruda.

### Con algunos elementos identificados podríamos limpiar la base de datos o modificarla para un óptimo análisis.

```python
#Renombrar columnas

apps= apps.rename(columns={'App': 'Aplicacion', 'Category':'Categoria', 'Rating':'clasificación', 'Reviews':'Puntuacion', 'Size':'Tamaño', 'Installs': 'Instalaciones', 'Type':'Tipo',
       'Price':'Precio', 'Content Rating':'Publico', 'Genres': 'Genero', 'Last Updated':'Ultima actualizacion', 'Current Ver':'Version act',
       'Android Ver': 'Version de android', 'Current Version': 'Version actual' , 'Creator':'Creador'})

#Eliminar columnas

#Borrar duplicados

apps = apps.drop_duplicates()

#Por criterio de NaN

#Borra filas con NaN
apps = apps.dropna(0, how='all')

#Borra columnas con NaN
apps = apps.dropna(1, how='all')

#eliminar filas especifica

#apps.drop(10995, axis=0, inplace=True)

#eliminar columna especifica

#apps.drop(['column name'], axis=1, inplace=True)

```

¿Qué transformaciones nos podrían gustar?

```python
#Saber la cantidad de descargas como numero y no por str

apps['Instalaciones']=apps['Instalaciones'].str.strip('+')

apps['Instalaciones']=apps['Instalaciones'].str.replace(',','')

apps['Instalaciones']=apps['Instalaciones'].astype('float64')

apps



```
