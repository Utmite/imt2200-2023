¬øComo podemos mostrar la relacion de dos variable?
Veamos  

```python
sns.relplot(
    data=df,
    x=df["Date and Time"].dt.hour, y="Count",
    hue=df["Date and Time"].dt.year
)
```

```python
sns.relplot(
    data=df,
    x=df["Date and Time"].dt.hour, y="Count",
     size=df["Date and Time"].dt.year
)
```

```python
sns.relplot(
    data=df,
    x=df["Date and Time"].dt.hour, y="Count",
     style=df["Date and Time"].dt.year
)
```

```python
sns.relplot(
    data=df,
    x=df["Date and Time"].dt.hour, y="Count",
    hue=df["Date and Time"].dt.year, kind='line'
)
```

```python
sns.relplot(
    data=df,
    x=df["Date and Time"].dt.hour, y="Count",
    hue=df["Date and Time"].dt.year, col=df["Date and Time"].dt.year
)
```

```python
df2= df[(df['Date and Time'].dt.year==2014) | (df['Date and Time'].dt.year==2013)]

```

```python
sns.relplot(
    data=df2,
    x=df2["Date and Time"].dt.hour, y="Count", col=df2["Date and Time"].dt.year
)
```

Vamos a usar la base de datos de üêß

```python
ping = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')
ping.head() # muestra primeros valores
```

```python
sns.boxplot(data=ping, y='bill_depth_mm', x='island')
```

image.png

Variable categoricas

```python
sns.catplot(data=ping, x="island", y="body_mass_g", hue='sex')
```

Juntemos plots

```python
sns.jointplot(data=ping, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

Hstograma

```python
sns.histplot(
    data=ping,
    x='body_mass_g'
)
```

Histograma

```python
sns.histplot(
    data=ping,
    x='body_mass_g',stat='probability'
)
```

Histograma

```python
sns.histplot(
    data=ping,
    x='body_mass_g',stat='density'
)
```

Hostograma

```python
sns.histplot(
    data=ping,
    x='body_mass_g',element='poly'
)
```

Mapa de calor

```python
sns.heatmap(ping.corr())
```

Violinplot

```python
sns.violinplot(data=ping, palette="Set3")
```

image.png

```python
sns.violinplot(data=ping[['bill_depth_mm','bill_length_mm','flipper_length_mm']], palette="Set3")
```

```python
sns.violinplot(data=ping['body_mass_g'], palette="Set3")
```
