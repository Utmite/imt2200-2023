# MANEJO DE STRINGS

Cuando trabajamos con datos, esta informaci√≥n nos puede llegar en diferentes formatos, por ejemplo, si guardamos la informacion de nombres de usuarios los guardamos como Strings.

Imaginemos que creamos una pagina web, donde la persona ingresa su nombre, pero se nos olvida indicarle que formato debe estar escrito (si todo en minusculas, may√∫sculas o una combinaci√≥n en eso)

```python
u = 'AMELIA LINDA'
l = 'amelia linda'
ul = 'AmelIa LinDa'

```

Nos podria llegar en estos tres formatos.

Consultemos que formato tienen los strings

```python
print(f'{u} es miniscula {u.islower()}')
print(f'{u} es mayuscula {u.isupper()}')

print(f'{l} es miniscula {l.islower()}')
print(f'{l} es mayuscula {l.isupper()}')

print(f'{ul} es miniscula {ul.islower()}')
print(f'{ul} es mayuscula {ul.isupper()}')
```

¬øC√≥mo podemos cambiar de formato a cada uno de los strings?

```python
#Queremos que sea todo en min√∫sculas
u.lower()
ul.lower()

#Queremos que sea todo en may√∫scula
l.upper()
ul.upper()

#Queremos que el caracter inicial sea may√∫scula

u.capitalize()

#Formato t√≠tulo

u.title()
```

# Hemos hecho una busqueda exhaustiva de distintos problemas e incoherencias que presenta el DataFrame. Solucionaremos cada uno de estos problemas mediante limpieza de datos.

```python
df_movies_clean = df_movies.copy()
```

- Eliminamos las columnas que tengan cierta cantidad de valores *nulos*.

```python
# para este ej p=60%
p = 0.6
thresh = int((1-p)*df_movies_clean.shape[0])
df_movies_clean.dropna(axis=1, thresh=thresh, inplace=True)

df_movies_clean.head()
```

- Eliminamos la columna 'ONE-LINE'

```python
df_movies_clean.drop(columns=['ONE-LINE'], inplace=True)
df_movies_clean
```

- Modificamos el nombre de la columna *RunTime*, para que quede en may√∫sculas:

```python
df_movies_clean.rename(columns={'RunTime' : 'RUNTIME'}, inplace=True)
df_movies_clean.head()
```

- Eliminamos las filas duplicadas

```python
print(f"Antes : {len(df_movies_clean)} filas")
df_movies_clean.drop_duplicates(inplace=True)
print(f"Despu√©s : {len(df_movies_clean)} filas")
```

- Eliminamos los saltos de l√≠nea en *GENRE* y *STARS*, y las ',' en *VOTES*

```python
df_movies_clean['GENRE'] = df_movies_clean['GENRE'].str.replace('\n', '')
df_movies_clean['STARS'] = df_movies_clean['STARS'].str.replace('\n', '')
df_movies_clean['VOTES'] = df_movies_clean['VOTES'].str.replace(',', '')

df_movies_clean.head()
```

- Llenemos los datos vac√≠o seg√∫n la columna

```python
values = {'YEAR' : 'NO INFORMATION', 'GENRE' : 'NO INFORMATION', 'RATING' : 0, 'STARS' : 'NO INFORMATION', 'VOTES' : 0, 'RUNTIME' : 0}
df_movies_clean.fillna(value=values, inplace=True)
df_movies_clean.head()
```

- Transformamos los datos de la columna *VOTES*, de *object* a *int64*

```python
df_movies_clean['VOTES'] = df_movies_clean['VOTES'].astype('int64')

df_movies_clean.info()
```
